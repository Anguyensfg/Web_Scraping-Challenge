```python
# Import Splinter, BeautifulSoup, Pandas and Webdriver

from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pymongo
```

```python
# Set the executable path and initialize Splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the red planets website

url = 'https://redplanetscience.com/'
browser.visit(url)

# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
# Convert the browser html to a soup object

html = browser.html
news_soup = bs(html, 'html.parser')

slide_elem = news_soup.select_one('div.list_text')
```

```python
# Display Results

slide_elem.find('div', class_='content_title')
```

```python
# Use the parent element to find the first a tag and save it as `news_title`

news_title = slide_elem.find('div', class_='content_title').get_text()
news_title
```

```python
# Use the parent element to find the preview text

news_p = slide_elem.find('div', class_='article_teaser_body').get_text()
news_p
```

```python
#Extract data from website to view 

print(news_soup.prettify())
```

```python
# Close the browser 

browser.quit()
```

```python
# Set the executable path and initialize Splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the red planets website

url = 'https://redplanetscience.com/'
browser.visit(url)

# Optional delay for loading the page

browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
# results are returned as an iterable list
results = news_soup.find_all('div', class_="list_text")
```

```python
# Loop through returned results
for result in results:
    try:
        # Select and return title
        title = result.find("div", class_="content_title").text
        # Select and return preview text
        preview = result.find("div", class_="article_teaser_body").text
        

        # Print title and preview text
        if (title and preview ):
            print('-------------')
            print(title)
            print(preview)
            
    except AttributeError as TL:
        print(TL)
        
```
